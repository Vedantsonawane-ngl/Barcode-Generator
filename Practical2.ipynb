1. writing to CSV

```python
import csv

# Field names
f =['Name','Branch','Year','CGPA']

# Data rows of CSV file
r = [
    ['Nikhil','COE','2','9.0'],
    ['Sanchit','COE','2','9.1'],
    ['Aditya','IT','2','9.3'],
    ['Sagar','SE','1','9.5'],
    ['Prateek','MCE','3','7.8'],
    ['Sahil','EP','2','9.1'],
]

# Name of the CSV file
fn ="student_records.csv"

# Writing to CSV file
with open(fn,'w') as csvfile:
    # Create a CSV writer object
    csvwriter = csv.writer(csvfile)

    # Writing the field names
    csvwriter.writerow(f)

    # Writing the data rows
    csvwriter.writerows(r)
```

converting .csv file to.json file format

```python
#converting .csv file to .json file format

import pandas as pd
import os

# Read CSV
df = pd.read_csv('student_records.csv')

# Convert to JSON
df.to_json('student_records.json',orient="records",indent=4)

print("Work perform succesfully!!")

#check correct file Extension

file_name = "student_records.json"
file_extension = os.path.splitext(file_name)[1]
```
